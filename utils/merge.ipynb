```python
#获取当前文件夹下所有文件名
for root, dirs, files in os.walk(r"E:\kaggle\newsclassification\正式比赛用的代码\1", topdown=False):
    for name in files:
        path = PurePath(name)
        csvlist.append(path.name)
```

```python
test_pred = np.zeros((50000, 1), int) 
test_pred
```

```python
for csv in csvlist:
    dirpath = Path("./1")
    filepath = dirpath.joinpath(csv)
    print(filepath)
    df = pd.read_csv(filepath)
    test_pred = np.column_stack((test_pred, df.values))
```
